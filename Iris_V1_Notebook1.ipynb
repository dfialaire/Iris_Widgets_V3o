```python
# -*- coding: utf-8 -*-
"""
Created on Sat Jun  6 21:20:18 2020

@author: dfial
"""
# -*- coding: utf-8 -*-
from tkinter import filedialog
from tkinter import *
from tkinter import messagebox
import pickle
import os
import sys, time
Lieu = os.getcwd()
import numpy as np
import cv2
from PIL import Image
from pylab import *
import matplotlib
from matplotlib.pyplot import *
import numpy
import statistics
from itertools import chain
import imageio
import webbrowser
###################################################
def sauvegard_image(matrice):
    # os.chdir(path)
#    scipy.misc.imageio.imwrite(name_matric,matric)
    #imageio.imwrite(name, matric)
    name_format="CaptWeb.jpg"
    imageio.imwrite(name_format, matrice[:, :])
#####################################################
#########

#########
font = cv2.FONT_HERSHEY_SIMPLEX  # FONT_HERSHEY_SIMPLEX  # FONT_HERSHEY_PLAIN
inverse=0
x2=274#240
y2=373#320
#############################################
#######################################################################
def OuvreDialogSave():
    # asksaveasfilename  # askopenfilename
    root = Tk()
    root.filename =  filedialog.asksaveasfilename(initialdir = "/",title = "Select file",filetypes = (("pickle files","*.pickle"),("all files","*.*")))
    root.destroy()
    texte=root.filename+".pickle"
    return texte
###########################
def OuvreDialogOpen():
    # asksaveasfilename  # askopenfilename
    root = Tk()
    root.filename =  filedialog.askopenfilename(initialdir = "/",title = "Select file",filetypes = (("pickle files","*.pickle"),("all files","*.*")))
    root.destroy()
    texte=root.filename
    return texte
##########################################
def Enregistrement(Saverootfilename,New_list):
    pickle_out=open(Saverootfilename,"wb+")
    pickle.dump(New_list,pickle_out)
    pickle_out.close()
###########################################
def SauveParametres(New_list):
    Saverootfilename=OuvreDialogSave()
    Enregistrement(Saverootfilename,New_list)
###########################################
def LireParametres():
    Openrootfilename=OuvreDialogOpen()
    pickle_in=open(Openrootfilename,"rb")
    After_Exist_List=pickle.load(pickle_in)
    pickle_in.close()
    Exist_list=After_Exist_List
    return Exist_list
###########################################
###########################################
def Souris(event,x,y,flags,param):
    global y2,x2
    if event == cv2.EVENT_LBUTTONDOWN:
        #cv2.circle(img,(x,y),5,(255,0,0),-1)
        y2 = x
        x2 = y
def trackChaned(x):
    pass
def fonction_p(img, lower_test, upper_test,inverse,w,wh, Nb_sup_wh, mode,List_info_paquets,diam_imp):
    modif=adjust_brightness(img, gamma)
    img=modif
    hauteur, largeur = img.shape[:2]
    # sortie d'image avec contours
    image_contours = img.copy()
    # Détection des colonies 
    colors = ['colonies']
    for color in colors:
        # copie de l'image originale
        image_to_process = img.copy()
        # initialisation du compteur
        counter[color] = 0
        # Dépôt des marges (GBR) de détection des couleurs dans un NumPy arrays
        if color == 'colonies':
            lower = np.array(lower_test)
            upper = np.array(upper_test)
        if inverse==1:
            image_to_process = (255-image_to_process)
        else:
            ok=2
        # Recherche des couleurs spécifiées dans les marges dans un masque
        image_mask2 = cv2.cvtColor(image_to_process, cv2.COLOR_BGR2HSV)
        image_mask = cv2.inRange(image_to_process, lower, upper)
        # application du masque
        image_res = cv2.bitwise_or(image_to_process, image_to_process, mask = image_mask)
        ## Transformation de l'image en nuance de gris et flouttage Gaussien
        image_gray = cv2.cvtColor(image_res, cv2.COLOR_BGR2GRAY)
        image_gray = cv2.GaussianBlur(image_gray, (5, 5), 0)
        # Détection des contours, et résolution des imperfections
        image_edged = cv2.Canny(image_gray, 50, 100)
        image_edged = cv2.dilate(image_edged, None, iterations=1)
        image_edged = cv2.erode(image_edged, None, iterations=1)
        
        if mode=="isole":
            cnts, hierarchy = cv2.findContours(image_edged.copy(), cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)
            cnts = cv2.findContours(image_edged.copy(), cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)
            
            
        else:
            cnts, hierarchy = cv2.findContours(image_edged.copy(), cv2.RETR_TREE, cv2.CHAIN_APPROX_NONE)
            cnts = cv2.findContours(image_edged.copy(), cv2.RETR_TREE, cv2.CHAIN_APPROX_NONE)
            #cnts = cv2.findContours(image_edged.copy(), cv2.RETR_CCOMP, cv2.CHAIN_APPROX_NONE)
            ###cnts = cv2.findContours(image_edged.copy(), cv2.RETR_LIST, cv2.CHAIN_APPROX_NONE)
            
            
            #cnts_tot = cv2.findContours(image_edged.copy(), cv2.RETR_LIST, cv2.CHAIN_APPROX_NONE) ##
            #cnts_tour = cv2.findContours(image_edged.copy(), cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)
            #cnts = cnts_tot - cnts_tour


            


        cnts = cnts[0] if len(cnts) == 2 else cnts[1]
    #####################################
        if mode=="tapis":
            if hierarchy is None:
                ko=2
            else:
                hierar = list(chain.from_iterable(hierarchy))
                lenhierar=len(hierar)
                listi=[]
                for xc in range(0, lenhierar):
                    xcv=hierar[xc]
                    xcvb=hierar[xc][3]
                    if xcvb==-1:
                        listi.append(xc)
    ####
                listi=sorted(listi, reverse=True)
                #print(listi)
                leny=len(listi)
                for a in range(0,leny):
                    u=listi[a]
                    del hierar[u]
                    del cnts[u]
    #####################################
        Aires=[]
        compte=1
        Aire_col_imp=int(10*(pow(diam_imp,2.0627)))  ##float
        Nb_sup_wh=0
        for c in cnts:
        
            # Si le contour n'est pas suffisamment large, l'ignorer : C'est un paramètre à modifier pour les tailles de colonies
            ###if Aire_col_imp<1:
                ###yes=2
            ###else:
            if cv2.contourArea(c) < w or cv2.contourArea(c)>wh:   ##1  #5
                if cv2.contourArea(c)>wh:
                    aire_compt=cv2.contourArea(c)
                    infos=(compte,aire_compt)
                    compte+=1
                    List_info_paquets.append(infos)
                    Nb_sup_wh=Nb_sup_wh+1
                    #print("Nb_sup_wh ligne 117=",Nb_sup_wh)
                continue

            if cv2.contourArea(c)>Aire_col_imp:
                aire_compt=cv2.contourArea(c)
                infos=(compte,aire_compt)
                compte+=1
                List_info_paquets.append(infos)
                Nb_sup_wh=Nb_sup_wh+1
                #print("Nb_sup_wh ligne 126=",Nb_sup_wh)
                #print("List_info_paquets ligne 127=",List_info_paquets)
##############
            Aire=cv2.contourArea(c)
            Aires.append(Aire)
            # Création de la coqie convexe du contour
            hull = cv2.convexHull(c)
            if color == 'colonies':
                # Marquer le contour en vert
                cv2.drawContours(image_contours,[hull],0,coul,1) #0,0,255 # 204,102,0
            counter[color] += 1
            # Décocher le commentaire ci-dessous pour écrire, sur l'image, les TAILLES de colonies détectées
            if affiche_Aire==1:
                cv2.putText(image_contours, "{:.0f}".format(cv2.contourArea(c)), (int(hull[0][0][0]), int(hull[0][0][1])), cv2.FONT_HERSHEY_SIMPLEX, 0.65, (0, 0, 0), 2)
            else:
                ko=1
        Len_Aires=len(Aires)
        if Len_Aires>0:
            Aire_Moy= int(statistics.mean(Aires) )
        else:
            Aire_Moy=0
        Aires=[]
    # Décocher le commentaire ci-dessous pour l'enregistrement de l'image finalisée
    #cv2.imwrite("Capture_denombrement.jpg",image_contours)     ######
    image_contours = image_contours.astype (np.uint8)
    return image_contours , counter[color],Nb_sup_wh,Aire_Moy,List_info_paquets
#####################################################    
def fonction_m(img,x,y, inverse): 
    p_unmin=255
    p_deuxmin=255
    p_troismin=255
    p_unmax=0
    p_deuxmax=0
    p_troismax=0
    if inverse==1:
        img = (255-img)
    else:
        ok=2
    image = cv2.cvtColor(img, cv2.COLOR_BGR2HSV)      
    xpmin=x
    xpmax=x+6
    ypmin=y
    ypmax=y+6
    cible_img=image[xpmin:xpmax,ypmin:ypmax]
    for xp in range(xpmin,xpmax-1):
        for yp in range(ypmin,ypmax-1):
            p=img[xp,yp]
            Prep_pixel = np.uint8([[p]])
            hsv_pixel = cv2.cvtColor(Prep_pixel,cv2.COLOR_BGR2HSV)
            pi=hsv_pixel[0]
            p_un=pi[0][0]
            p_deux=pi[0][1]
            p_trois=pi[0][2]
            if p_un<p_unmin:
                p_unmin=p_un
            else:
                z=1
            if p_un>p_unmax:
                p_unmax=p_un
            else:
                z=1
            if p_deux<p_deuxmin:
                p_deuxmin=p_deux
            else:
                z=1
            if p_deux>p_deuxmax:
                p_deuxmax=p_deux
            else:
                z=1 
            if p_trois<p_troismin:
                p_troismin=p_trois
            else:
                z=1
            if p_trois>p_troismax:
                p_troismax=p_trois
            else:
                z=1 
    #lower_test=[p_unmin,25,25]
    #lower_test=[p_unmin,p_deuxmin,p_troismin] ## cool
    lower_test=[p_unmin-32,p_deuxmin,p_troismin]  ###
    #upper_test=[p_unmax,255,255]
    #upper_test=[p_unmax,p_deuxmax,p_troismax]  ## cool
    upper_test=[p_unmax+32,p_deuxmax,p_troismax]  ###
    #print("lower=",lower_test)
    #print("upper=",upper_test)
    return lower_test, upper_test
####################################################
def zoom_photo2(img,x2,y2):
    new_img=img[x2-20:x2+20,y2-30:y2+30]#[y+47:y+87,x-13:x+47]#[y+30:y+70,x-10:x+50]#[y+30:y+90,x-10:x+30]#[y:y+60,x+40:x+80]   ##[280:360,210:270]
    img = cv2.resize(new_img,(640,480))
    return img
####################################################
def cible_petite(img,xc,yc):
    xcmin=xc-3
    xcmax=xc+3
    ycmin=yc-3
    ycmax=yc+3
    for x in range(xcmin,xcmax):
        img[ycmin,x]=[0,0,255]
    for x in range(xcmin,xcmax):
        img[ycmax,x]=[0,0,255]

    for y in range(ycmin,ycmax):
        img[y,xcmin]=[0,0,255]
    for y in range(ycmin,ycmax):
        img[y,xcmax]=[0,0,255]
    return img
####################################################
def cible(img,yc,xc):
    xcmax=xc+64
    ycmax=yc+64
    for x in range(xc,xcmax):
        img[yc,x]=[0,255,0]
    for x in range(xc,xcmax):
        img[ycmax,x]=[0,255,0]

    for y in range(yc,ycmax):
        img[y,xc]=[0,255,0]
    for y in range(yc,ycmax):
        img[y,xcmax]=[0,255,0]
    return img
##################################################
def affich_image_array_pil(Matrice_Name):
    image = numpy.array(Matrice_Name)
    gray()
    # montrer l'image
    imshow(image)
#############################################
def adjust_brightness(img, brightness_factor):
    """Adjust brightness of an Image.
    Args:
        img (numpy ndarray): numpy ndarray to be adjusted.
        brightness_factor (float):  How much to adjust the brightness. Can be
            any non negative number. 0 gives a black image, 1 gives the
            original image while 2 increases the brightness by a factor of 2.
    Returns:
        numpy ndarray: Brightness adjusted image.
    """
    table = np.array([ i*brightness_factor for i in range (0,256)]).clip(0,255).astype('uint8')
    if img.shape[2]==1:
        return cv2.LUT(img, table)[:,:,np.newaxis]
    else:
        return cv2.LUT(img, table) 
##################################################
def transform_taille(img):
    img.thumbnail((640,480))
    return img

####################################################
def zoom_photo(img,zoom):
    largeur = img.shape[0]
    hauteur = img.shape[1]   
    new_largeur=int(largeur/zoom)
    new_hauteur=int(hauteur/zoom)
    new_img = numpy.empty((new_largeur, new_hauteur, 3), dtype = numpy.uint8)
    for l in range (new_largeur):
        for h in range (new_hauteur):
            new_img[l,h]=img[l,h]
    img = cv2.resize(new_img,(640,480))  
    return img
#############################################
def affich(img,titre):
    cv2.imwrite(titre,img) ## modif
    Affich_fin = Image.open(titre)
    affich_image_array_pil(Affich_fin)
    title(titre)
    show()
#########################################
  # dictionnaire de comptage des colonies
counter = {}
##################################
# ça commence ici ################
##################################
cam=cv2.VideoCapture(0)

cv2.namedWindow("\"_ Iris _\" : Denombrement de colonies (David Fialaire _ Juin 2020 _ Version 1.0) :",1)
cv2.resizeWindow("\"_ Iris _\" : Denombrement de colonies (David Fialaire _ Juin 2020 _ Version 1.0) :", (1900,700))
cv2.moveWindow("\"_ Iris _\" : Denombrement de colonies (David Fialaire _ Juin 2020 _ Version 1.0) :", 0,0)
cv2.namedWindow("Tutoriels :",1)
cv2.resizeWindow("Tutoriels :", (320,50))
cv2.moveWindow("Tutoriels :", 650,485)
cv2.namedWindow("Versions a telecharger ? :",1)
cv2.resizeWindow("Versions a telecharger ? :", (320,50))
cv2.moveWindow("Versions a telecharger ? :", 650,565)
cv2.resizeWindow("Modes :", (500,300))
cv2.namedWindow("Modes :",1)
cv2.resizeWindow("Modes :", (320,80))
cv2.moveWindow("Modes :", 973,413)
cv2.namedWindow("Fonction de la taille :",1)
cv2.resizeWindow("Fonction de la taille :", (320,80))
cv2.moveWindow("Fonction de la taille :", 650,370)
cv2.namedWindow("Donnees Parametres :",1)
cv2.resizeWindow("Donnees Parametres :", (320,120))
cv2.moveWindow("Donnees Parametres :", 973,262)
cv2.namedWindow("Type de calibration :",1)
cv2.resizeWindow("Type de calibration :", (320,85))
cv2.moveWindow("Type de calibration :", 973,35)
cv2.namedWindow("Controles",1)  # 1 # cv2.WINDOW_KEEPRATIO
cv2.namedWindow('Controle sur Aire imposee',1)
cv2.resizeWindow("Controles", (320,300))
cv2.moveWindow("Controles", 650,35)
cv2.resizeWindow('Controle sur Aire imposee', (320,80))
cv2.moveWindow('Controle sur Aire imposee', 973,150)
cv2.namedWindow("Webcam :",1)
cv2.resizeWindow("Webcam :", (320,80))
cv2.moveWindow("Webcam :", 973,525)
#        cv2.namedWindow(window_name,
#                        cv2.WINDOW_KEEPRATIO | cv2.WINDOW_GUI_NORMAL)
#        cv2.resizeWindow(window_name, *size_wh)    ### size_wh: window size (w, h)

gamma=1
zoom=1
x = 306
y = 176
x2=274#240
y2=373#320
p_unmin=255
p_deuxmin=255
p_troismin=255
p_unmax=0
p_deuxmax=0
p_troismax=0
zoom2=0
rayon=150
Comptage_min=400
Comptage_max=0
w=1
wh=300
Nb_List=0
Nb_List_max=10
List_max=[]

Nb_List_min=0
List_min=[]
mode="isole"
Nb_sup_wh=0
List_info_paquets=[]
ListeNbColPlus=[]
cl=0
cl_new=0
fo=0
fo_new=0
weballum=0
#Aire_col_imp=73
#diam_imp=round((Aire_col_imp/10)**(1/2.0627))
#"Type de calibration :"

cv2.createTrackbar("-->Site", "Versions a telecharger ? :",0,1,trackChaned)
cv2.createTrackbar("Direct :", "Webcam :",0,1,trackChaned)
cv2.createTrackbar("Capture:", "Webcam :",0,1,trackChaned)

cv2.createTrackbar("Tuto :", "Tutoriels :",0,9,trackChaned)
cv2.createTrackbar("isolees", "Modes :",1,1,trackChaned)
cv2.createTrackbar("paquets", "Modes :",0,1,trackChaned)
cv2.createTrackbar("Diam.inf", "Fonction de la taille :",0,11,trackChaned)
cv2.createTrackbar("Diam.sup", "Fonction de la taille :",11,11,trackChaned)
cv2.createTrackbar("Charger", "Donnees Parametres :",0,1,trackChaned)
cv2.createTrackbar("Sauver", "Donnees Parametres :",0,1,trackChaned)
cv2.createTrackbar("Reset", "Donnees Parametres :",0,1,trackChaned)
cv2.createTrackbar("1=Colonie", "Type de calibration :",1,1,trackChaned)
cv2.createTrackbar("1=Milieu", "Type de calibration :",0,1,trackChaned)
cv2.createTrackbar("1=Exit", "Controles",0,1,trackChaned)
cv2.createTrackbar("Couleur", "Controles",0,2,trackChaned)
#cv2.createTrackbar("x=", "Controles",373,640,trackChaned)  #320
#cv2.createTrackbar("y=", "Controles",274,480,trackChaned)  #240 "Controles"
cv2.createTrackbar("Affich=", "Controle sur Aire imposee",0,1,trackChaned)
cv2.createTrackbar("+Clair=", "Controles",0,25,trackChaned)
cv2.createTrackbar("+Fonce=", "Controles",0,25,trackChaned)
cv2.createTrackbar("Big=", "Controles",0,1,trackChaned)
cv2.createTrackbar("Small=", "Controles",0,1,trackChaned)
cv2.createTrackbar("Zoom =", "Controles",0,1,trackChaned)
#cv2.createTrackbar("Aire_Colonie", "Controles",1,300,trackChaned)
cv2.createTrackbar("Aire=", "Controle sur Aire imposee",0,300,trackChaned)

#Controle sur Aire imposee
while(True): 
########################
 
###################
    lire=cv2.getTrackbarPos("Charger", "Donnees Parametres :")
    if lire==1:
        gamma=1
        zoom=1
        x = 306
        y = 176
        x2=274#240
        y2=373#320
        p_unmin=255
        p_deuxmin=255
        p_troismin=255
        p_unmax=0
        p_deuxmax=0
        p_troismax=0
        zoom2=0
        rayon=150
        Comptage_min=400
        Comptage_max=0
        w=1
        wh=300
        Nb_List=0
        Nb_List_max=10
        List_max=[]

        Nb_List_min=0
        List_min=[]
        mode="isole"
        Nb_sup_wh=0
        List_info_paquets=[]
        ListeNbColPlus=[]
        cl=0
        cl_new=0
        fo=0
        fo_new=0
        weballum=0
        inverse=0
        cv2.setTrackbarPos("Charger", "Donnees Parametres :",0)
        listo=LireParametres()
        #diam_imp,Aire_col_imp,affiche_Aire,couleur,inv,inv2,diam_col_sup,diam_col_inf,tuto,webi,gamma,zoom,x2,y2,p_unmin,p_deuxmin,p_troismin,p_unmax,p_deuxmax,p_troismax,zoom2,rayon,Comptage_min,Comptage_max,w,wh,Nb_List,Nb_List_max,List_max,Nb_List_min,List_min,mode,Nb_sup_wh,List_info_paquets,ListeNbColPlus,cl,cl_new,fo,fo_new,weballum,inverse=listo[0],listo[1],listo[2],listo[3],listo[4],listo[5],listo[6],listo[7],listo[8],listo[9],listo[10],listo[11],listo[12],listo[13],listo[14],listo[15],listo[16],listo[17],listo[18],listo[19],listo[20],listo[21],listo[22],listo[23],listo[24],listo[25],listo[26],listo[27],listo[28],listo[29],listo[30],listo[31],listo[32],listo[33],listo[34],listo[35],listo[36],listo[37],listo[38],listo[39],listo[40]
        Aire_col_imp,webi,tuto,mode,diam_col_inf,diam_col_sup,inverse,couleur,affiche_Aire,cl,fo,x2,y2=listo[0],listo[1],listo[2],listo[3],listo[4],listo[5],listo[6],listo[7],listo[8],listo[9],listo[10],listo[11], listo[12]
        print("Après Lire : inv=",inv," et inv2=",inv2, " et inverse =",inverse)
        cv2.setTrackbarPos("Direct :", "Webcam :",webi)#
        cv2.setTrackbarPos("Tuto :", "Tutoriels :",tuto)#
        if mode=="isole":
            cv2.setTrackbarPos("isolees", "Modes :",1)
            cv2.setTrackbarPos("paquets", "Modes :",0)
        else:
            cv2.setTrackbarPos("isolees", "Modes :",0)
            cv2.setTrackbarPos("paquets", "Modes :",1)            
        cv2.setTrackbarPos("Diam.inf", "Fonction de la taille :",diam_col_inf)
        cv2.setTrackbarPos("Diam.sup", "Fonction de la taille :",diam_col_sup)
        cv2.setTrackbarPos("Charger", "Donnees Parametres :",0)
        cv2.setTrackbarPos("Sauver", "Donnees Parametres :",0)
        cv2.setTrackbarPos("Reset", "Donnees Parametres :",0)
        #cv2.setTrackbarPos("1=Colonie", "Type de calibration :",inv)#
        #cv2.setTrackbarPos("1=Milieu", "Type de calibration :",inv2)
        cv2.setTrackbarPos("1=Exit", "Controles",0)
        cv2.setTrackbarPos("Couleur", "Controles",couleur)#
        cv2.setTrackbarPos("Affich=", "Controle sur Aire imposee",affiche_Aire)#
        cv2.setTrackbarPos("+Clair=", "Controles",cl)
        cv2.setTrackbarPos("+Fonce=", "Controles",fo)
        cv2.setTrackbarPos("Big=", "Controles",0)
        cv2.setTrackbarPos("Small=", "Controles",0)
        cv2.setTrackbarPos("Zoom =", "Controles",0)
        Aire_col_imp=int(Aire_col_imp)
        cv2.setTrackbarPos("Aire=", "Controle sur Aire imposee",Aire_col_imp)#
        if inverse==0:
            cv2.setTrackbarPos("1=Colonie", "Type de calibration :",1)#
            cv2.setTrackbarPos("1=Milieu", "Type de calibration :",0)   
        else:
            cv2.setTrackbarPos("1=Colonie", "Type de calibration :",0)#
            cv2.setTrackbarPos("1=Milieu", "Type de calibration :",1)
########################

    Res=cv2.getTrackbarPos("Reset", "Donnees Parametres :")
    if Res==1:
        gamma=1
        zoom=1
        x = 306
        y = 176
        x2=274#240
        y2=373#320
        p_unmin=255
        p_deuxmin=255
        p_troismin=255
        p_unmax=0
        p_deuxmax=0
        p_troismax=0
        zoom2=0
        rayon=150
        Comptage_min=400
        Comptage_max=0
        w=1
        wh=300
        Nb_List=0
        Nb_List_max=10
        List_max=[]

        Nb_List_min=0
        List_min=[]
        mode="isole"
        Nb_sup_wh=0
        List_info_paquets=[]
        ListeNbColPlus=[]
        cl=0
        cl_new=0
        fo=0
        fo_new=0
        weballum=0
        inverse=0
        cv2.setTrackbarPos("Direct :", "Webcam :",0)
        cv2.setTrackbarPos("Tuto :", "Tutoriels :",0)
        cv2.setTrackbarPos("isolees", "Modes :",1)
        cv2.setTrackbarPos("paquets", "Modes :",0)
        cv2.setTrackbarPos("Diam.inf", "Fonction de la taille :",0)
        cv2.setTrackbarPos("Diam.sup", "Fonction de la taille :",11)
        cv2.setTrackbarPos("Charger", "Donnees Parametres :",0)
        cv2.setTrackbarPos("Sauver", "Donnees Parametres :",0)
        cv2.setTrackbarPos("Reset", "Donnees Parametres :",0)
        cv2.setTrackbarPos("1=Colonie", "Type de calibration :",1)
        cv2.setTrackbarPos("1=Milieu", "Type de calibration :",0)
        cv2.setTrackbarPos("1=Exit", "Controles",0)
        cv2.setTrackbarPos("Couleur", "Controles",0)
        cv2.setTrackbarPos("Affich=", "Controle sur Aire imposee",0)
        cv2.setTrackbarPos("+Clair=", "Controles",0)
        cv2.setTrackbarPos("+Fonce=", "Controles",0)
        cv2.setTrackbarPos("Big=", "Controles",0)
        cv2.setTrackbarPos("Small=", "Controles",0)
        cv2.setTrackbarPos("Zoom =", "Controles",0)
        cv2.setTrackbarPos("Aire=", "Controle sur Aire imposee",0)
       
        ########   webbrowser.open('http://www.python.org')
    site=cv2.getTrackbarPos("-->Site", "Versions a telecharger ? :")
    if site==1:
        cv2.setTrackbarPos ("-->Site", "Versions a telecharger ? :",0)
        webbrowser.open('https://biotech-online.pagesperso-orange.fr/Soutien%20TP/Iris.htm')
    else:
        ok=1
    webi=cv2.getTrackbarPos("Direct :", "Webcam :")
    capt=cv2.getTrackbarPos("Capture:", "Webcam :")
    zoom2=cv2.getTrackbarPos("Zoom =", "Controles")
    tuto=cv2.getTrackbarPos("Tuto :", "Tutoriels :")
    diam_col_inf=cv2.getTrackbarPos("Diam.inf", "Fonction de la taille :")
    w=float(10*(pow(diam_col_inf,2.0627)))
    diam_col_sup=cv2.getTrackbarPos("Diam.sup", "Fonction de la taille :")
    wh=float(10*(pow(diam_col_sup,2.0627)))
    agr=cv2.getTrackbarPos("Big=", "Controles")
    dim=cv2.getTrackbarPos("Small=", "Controles")
    if agr==1:
        cv2.setTrackbarPos ("Big=", "Controles",0)
        rayon+=2
    if dim==1:
        cv2.setTrackbarPos ("Small=", "Controles",0)
        rayon-=2
    cl=cv2.getTrackbarPos("+Clair=", "Controles")
    fo=cv2.getTrackbarPos("+Fonce=", "Controles")
    if cl>cl_new:
        if fo==0:
            cl_new+=1
        else:
            cv2.setTrackbarPos ("+Clair=", "Controles",cl_new)
            fo_new=fo_new-1
            cv2.setTrackbarPos ("+Fonce=", "Controles",fo_new)

    if fo>fo_new:
        if cl==0:
            fo_new+=1
        else:
            cv2.setTrackbarPos ("+Fonce=", "Controles",fo_new)
            cl_new=cl_new-1
            cv2.setTrackbarPos ("+Clair=", "Controles",cl_new)
    if fo==0:
        gamma=float(cl/10)+1
    if cl==0:
        gamma=1-float(fo/10)
###############################################""
    
    couleur=cv2.getTrackbarPos("Couleur", "Controles")
    if couleur==0:
        coul=(255,0,0)
    elif couleur==1:
        coul=(0,255,0)
    elif couleur==2:
        coul=(0,0,255)
    affiche_Aire=cv2.getTrackbarPos("Affich=", "Controle sur Aire imposee")
    Aire_col_imp=cv2.getTrackbarPos("Aire=", "Controle sur Aire imposee")
    #y2=cv2.getTrackbarPos("x=", "Controles")
    #x2=cv2.getTrackbarPos("y=", "Controles")
#################################################    
    inv=cv2.getTrackbarPos("1=Colonie", "Type de calibration :")
    #print("inv=",inv)
    inv2=cv2.getTrackbarPos("1=Milieu", "Type de calibration :")
    #print("inv2=",inv2)
    if inv2==1: ## inv2=1
        if inv==1:  #  inv2=1 et inv=1
            if inverse==0:  #  inv2=1 et inv=1 et inverse=0
                #print("inverse=",inverse)
                inverse=1
                cv2.setTrackbarPos ("1=Colonie", "Type de calibration :",0)
                inv=cv2.getTrackbarPos("1=Colonie", "Type de calibration :")

                #print("yo1")
                #print("inverse_yo1=",inverse)
            else: # inv2=1 et inv=1 et inverse=1
                #print("inverse=",inverse)
                inverse=0
                cv2.setTrackbarPos ("1=Milieu", "Type de calibration :",0)
                inv2=cv2.getTrackbarPos("1=Milieu", "Type de calibration :")

                #print("yo2")
                #print("inverse_yo2=",inverse)
        else:  #  inv2=1 et inv=0
            #print("inverse=",inverse)
            inverse=1
            #print("inverse=",inverse)
            cv2.setTrackbarPos ("1=Colonie", "Type de calibration :",0)
            inv=cv2.getTrackbarPos("1=Colonie", "Type de calibration :")
            cv2.setTrackbarPos ("1=Milieu", "Type de calibration :",1)
            inv2=cv2.getTrackbarPos("1=Milieu", "Type de calibration :")
            #print("yo3")
            #print("inverse_yo3=",inverse)
    else: ## inv2=0
        if inv==0:  ## inv2 =0 et inv =0
            if inverse==0:  ## inv2 =0 et inv =0  inverse=0
                inverse=1
                #print("Newinverse=",inverse)
                cv2.setTrackbarPos ("1=Milieu", "Type de calibration :",1)
                inv2=cv2.getTrackbarPos("1=Milieu", "Type de calibration :")
                #print("yaa1")
                #print("inverse_yya1=",inverse)
            else:   ## inv2 =0 et inv =0  inverse=1
                inverse=0
                #print("Newinverse=",inverse)
                cv2.setTrackbarPos ("1=Colonie", "Type de calibration :",1)
                inv=cv2.getTrackbarPos("1=Colonie", "Type de calibration :")
                #print("yaa2")
                #print("inverse_yaa2=",inverse)
        else:  ## inv=1 et inv2=0
            #print("inverse=",inverse)
            ###inverse=0
            ###print("inverse=",inverse)
            cv2.setTrackbarPos ("1=Colonie", "Type de calibration :",1)
            inv=cv2.getTrackbarPos("1=Colonie", "Type de calibration :")
            cv2.setTrackbarPos ("1=Milieu", "Type de calibration :",0)
            inv2=cv2.getTrackbarPos("1=Milieu", "Type de calibration :")
            #print("yaa3")
            #print("inverse_yaa3=",inverse)
##############################################################################
#################################################    222222
    inv=cv2.getTrackbarPos("isolees", "Modes :")
    #print("inv=",inv)
    inv2=cv2.getTrackbarPos("paquets", "Modes :")
    #print("inv2=",inv2)
    if inv2==1: ## inv2=1
        if inv==1:  #  inv2=1 et inv=1
            if mode=="isole":  #  inv2=1 et inv=1 et inverse=0
                #print("inverse=",inverse)
                mode="tapis"
                cv2.setTrackbarPos ("isolees", "Modes :",0)
                inv=cv2.getTrackbarPos("isolees", "Modes :")

                #print("yo1")
                #print("inverse_yo1=",inverse)
            else: # inv2=1 et inv=1 et inverse=1
                #print("inverse=",inverse)
                mode="isole"
                cv2.setTrackbarPos ("paquets", "Modes :",0)
                inv2=cv2.getTrackbarPos("paquets", "Modes :")

                #print("yo2")
                #print("inverse_yo2=",inverse)
        else:  #  inv2=1 et inv=0
            #print("inverse=",inverse)
            mode="tapis"
            #print("inverse=",inverse)
            cv2.setTrackbarPos ("isolees", "Modes :",0)
            inv=cv2.getTrackbarPos("isolees", "Modes :")
            cv2.setTrackbarPos ("paquets", "Modes :",1)
            inv2=cv2.getTrackbarPos("paquets", "Modes :")
            #print("yo3")
            #print("inverse_yo3=",inverse)
    else: ## inv2=0
        if inv==0:  ## inv2 =0 et inv =0
            if mode=="isole":  ## inv2 =0 et inv =0  inverse=0
                mode="tapis"
                #print("Newinverse=",inverse)
                cv2.setTrackbarPos ("paquets", "Modes :",1)
                inv2=cv2.getTrackbarPos("paquets", "Modes :")
                #print("yaa1")
                #print("inverse_yya1=",inverse)
            else:   ## inv2 =0 et inv =0  inverse=1
                mode="isole"
                #print("Newinverse=",inverse)
                cv2.setTrackbarPos ("isolees", "Modes :",1)
                inv=cv2.getTrackbarPos("isolees", "Modes :")
                #print("yaa2")
                #print("inverse_yaa2=",inverse)
        else:  ## inv=1 et inv2=0
            #print("inverse=",inverse)
            ###inverse=0
            ###print("inverse=",inverse)
            cv2.setTrackbarPos ("isolees", "Modes :",1)
            inv=cv2.getTrackbarPos("isolees", "Modes :")
            cv2.setTrackbarPos ("paquets", "Modes :",0)
            inv2=cv2.getTrackbarPos("paquets", "Modes :")
            #print("yaa3")
            #print("inverse_yaa3=",inverse)
##############################################################################222222
            
# =============================================================================
#     if inv2==1 and inv==1:
#         if inverse==0:
#             cv2.setTrackbarPos ("1=Colonie", "Type de calibration :",0)
#         else:
#             cv2.setTrackbarPos ("1=Milieu", "Type de calibration :",0)
#     if inv2==0 and inv==0:
#         if inverse==0:
#             cv2.setTrackbarPos ("1=Milieu", "Type de calibration :",1)
#         else:
#             cv2.setTrackbarPos ("1=Colonie", "Type de calibration :",1)           
#             
# =============================================================================


    quit=cv2.getTrackbarPos("1=Exit", "Controles")
    if quit==1:
        break

    diam_imp=round((Aire_col_imp/10)**(1/2.0627))
    
    sauve=cv2.getTrackbarPos("Sauver", "Donnees Parametres :")
    if sauve==1:
        cv2.setTrackbarPos("Sauver", "Donnees Parametres :",0)
        
        print("Avant sauvegarde : inv=",inv," et inv2=",inv2, " et inverse =",inverse)
        New_list=[Aire_col_imp,webi,tuto,mode,diam_col_inf,diam_col_sup,inverse,couleur,affiche_Aire,cl,fo,x2,y2]
        # [96.41797634403324, 0, 3, 'isole', 0, 11, 0, 0, 1, 0, 0, 274, 373]
        print("New_list=",New_list)
        print("Avant sauvegarde : diam_imp=",diam_imp," et Aire_col_imp=",Aire_col_imp, " et inverse =",inverse)
        SauveParametres(New_list)
        print("Après sauvegarde : inv=",inv," et inv2=",inv2, " et inverse =",inverse)
        print("Après sauvegarde : diam_imp=",diam_imp," et Aire_col_imp=",Aire_col_imp, " et inverse =",inverse)
########################
    #print("aire_col_imp ligne 582=",Aire_col_imp)
    #print("diam-imp ligne 582=",diam_imp)
    
    #print("Aire_col_imp=",Aire_col_imp)
    #print("diam_imp=",diam_imp)
    
    #if Aire_col_imp==0:
        #Aire_col_imp=1
        #diam_imp=round((Aire_col_imp/10)**(1/2.0627))

#########################
    
    
    
    #ret,img=cam.read()
    ########################################""
    if webi==0:
        if weballum==1:
            cam.release()
            weballum=0
            list_titre=["calib3.jpg","calib1.jpg","Bleu2c.jpg","test5.jpg","test4.jpg","VRBL3.jpg","Staph1.jpg","exo1.jpg","BleuViolet.jpg","CaptWeb.jpg"]
            titre=list_titre[tuto]
            pil_img=Image.open(titre) # test4.jpg  # calib1.jpg# Bleu2.jpg # test3_pix1.jpg  # Staph1.jpg # VRBL3.jpg
            img = numpy.array(pil_img)
        else:
            list_titre=["calib3.jpg","calib1.jpg","Bleu2c.jpg","test5.jpg","test4.jpg","VRBL3.jpg","Staph1.jpg","exo1.jpg","BleuViolet.jpg","CaptWeb.jpg"]
            titre=list_titre[tuto]
            pil_img=Image.open(titre) #TBX.jpg # test4.jpg  # calib1.jpg# Bleu2.jpg # test3_pix1.jpg  # Staph1.jpg # VRBL3.jpg
            img = numpy.array(pil_img)
    else:
        if capt==1:
            cam=cv2.VideoCapture(0)
            ret,img=cam.read()
            weballum=1
            capture=img
            sauvegard_image(capture)
            pil_img=Image.open("CaptWeb.jpg") 
            img = numpy.array(pil_img)
            cam.release()
            weballum=0
            cv2.setTrackbarPos("Direct :", "Webcam :",0)
            cv2.setTrackbarPos("Capture:", "Webcam :",0)
            cv2.setTrackbarPos("Tuto :", "Tutoriels :",9)
            webi=0

        else:
            cam=cv2.VideoCapture(0)
            ret,img=cam.read()
            weballum=1

    #####################################""
    modif=adjust_brightness(img, gamma)
    img=modif
    if zoom2==1:
        img=zoom_photo2(img,x2,y2)
        #img=cible(img,x,y)
        #cv2.imshow("En direct _ Video Webcam", img)
    else:
        ok=2
##############
    ## Formation du masque noir autour de la boîte
    mask_circle = np.zeros(img.shape[:2],np.uint8)
    cv2.circle(mask_circle,(320,240), rayon, 255, -1)
    img = cv2.bitwise_and(img, img, mask=mask_circle) 
##############
    img=cible_petite(img,y2,x2)
    lower_test,upper_test=fonction_m(img,x2, y2,inverse)
    img,comptage,Nb_sup_wh,Aire_Moy,List_info_paquets=fonction_p(img, lower_test, upper_test,inverse,w,wh,Nb_sup_wh, mode,List_info_paquets,diam_imp)
###################
    if Nb_List!=10:
        List_max.append(comptage)
        Nb_List+=1

    else:
        val=max(List_max)
        Comptage_max=val
        val_min=val-((20/10)*val)
        for elem in List_max:
            if elem<val_min:
                List_max.remove(elem)
        valo=min(List_max)
        Comptage_min=valo
        Nb_List=0
        List_max=[]


    moyenne=(Comptage_min+Comptage_max)/2
###################
    Texte1="On compte : "
    Texte2=" colonies."
    texte2b=" colonie."
    if comptage<2:
        Texte=str(Texte1)+str(comptage)+str(texte2b)
    else:
        Texte=str(Texte1)+str(comptage)+str(Texte2)
    Texte3="Calibration_couleur="
    if inverse==1:
        texte11="Milieu"
    else:
        texte11="Colonie"
    Texte4=str(Texte3)+str(texte11)
    if mode=="isole":
        Texte9=str("isolees")
    else:
        Texte9=str("rapprochees")
    Texte8="Mode : "+Texte9
    Taille_limit=int(w/150)
    #Texte10=str(w)+" < Aire de 'colonie(s)' comptee(s) < "+str(wh)
    diam_col_inf=round((w/10)**(1/2.0627))
    diam_col_sup=round((wh/10)**(1/2.0627))
    #diam_col_moy=round((Aire_Moy+770)/193.5,2)
    #Texte19=str(diam_col_inf)+" < Diametre de 'colonie(s)' comptee(s) < "+str(diam_col_sup)
    #Texte12="Nombre de paquets de colonies > wh = "+str(Nb_sup_wh)
    #Texte13="Diametre moyen impose des colonies en paquets : "+str(diam_imp)
########################################
    if Aire_col_imp>0:
        #print("Aire_col_imp ligne 678=",Aire_col_imp)
        Texte23="Aire imposee ="+str(Aire_col_imp)
        #print("Aire imposee ligne684=",Aire_col_imp)
        #print(type(Aire_col_imp))
        cv2.putText(img, Texte23 ,(5,290), font, 0.5, (0,0,255), 1);
        ListeAirePaquet=[]
        ################

        NbColIsol=comptage-Nb_sup_wh
        #print("Nb_sup_wh ligne 690=",Nb_sup_wh )
        for paquet in range (0,Nb_sup_wh):
            #print("paquet=",paquet)
            #print("List_info_paquets=")
            #print(List_info_paquets)
            Num_paquet=List_info_paquets[paquet][0]
            #print("Num_paquet=",Num_paquet)
            Aire_paquet=List_info_paquets[paquet][1]
            #print("Aire_paquet=",Aire_paquet)   
            Aire_col_imp_elev=float(Aire_col_imp+((23/100)*Aire_col_imp))  ##float
            #print("Aire_col_imp_elev=",Aire_col_imp_elev)
            Nb_col=round(int(Aire_paquet/Aire_col_imp_elev),0)  ##float
            #Texte17="Paquet:"+str(Num_paquet)+", Nb.col="+str(Aire_paquet)
            Texte17="Paquet:"+str(Num_paquet)+", Nb.col="+str(Nb_col)
            ListeNbColPlus.append(Nb_col)
            hauteur=378+(12*Num_paquet)
            ListeAirePaquet.append(Aire_paquet)
            cv2.putText(img, Texte17 ,(5,hauteur), font, 0.5, (0,0,255), 1);

            #print("List_info_paquets=",List_info_paquets)
        ############
        #Texte23="Aire imposee ="+str(Aire_col_imp)
        #cv2.putText(img, Texte23 ,(5,180), font, 0.5, (0,0,255), 1);
        SomColPlus=int(sum(ListeNbColPlus))
        Texte21="Nb col. isolee:"+str(NbColIsol)
        #print("NbColIsol ligne 717=",NbColIsol)
        #print(type(NbColIsol))
        #Texte20="Nb col. paquets en +:"+str(SomColPlus)+"."
        Texte20="Nb col. paquets en +:"+str(SomColPlus)
        #print("Nb col. paquets en + ligne 721)",SomColPlus)
        NbTotcol=NbColIsol+SomColPlus  ##float
        Ni=NbTotcol
        Texte22="Nombre total de colonies: "+str(Ni)
        #Texte22="Nombre total de colonies: "+str(NbTotcol)
        cv2.putText(img, Texte20 ,(5,340), font, 0.5, (0,0,255), 1);
        cv2.putText(img, Texte21 ,(5,315), font, 0.5, (0,0,255), 1);
        cv2.putText(img, Texte22 ,(5,93), font, 0.5, (0,0,255), 1);
        #cv2.putText(img, "{:.0f}".format(Texte22) ,(5,93), font, 0.5, (0,0,255), 1);


        ListeNbColPlus=[]
        Aire_col_imp=float(10*(pow(diam_imp,2.0627)))  ##float
        #print("diam_imp=",diam_imp)
        #print("Aire_col_imp de diamp_imp=",Aire_col_imp)
        SomeListeAirePaquet=sum(ListeAirePaquet)
        #SommColTot=round(float(sum(SomeListeAirePaquet/Aire_col_imp)),1)
        #if Aire_col_imp==0:
            #Aire_col_imp=1
        SommColTot=float((SomeListeAirePaquet/Aire_col_imp)*(75/100))  ##float
        #Texte18="Nb colonie(s) deduite(s) en + ="+str(SommColTot)
        #cv2.putText(img, Texte18 ,(5,410), font, 0.5, (255,255,255), 1);
        #############

        ##########
        Nb_sup_wh=0
    List_info_paquets=[]
    cv2.putText(img, Texte8 ,(260,30), font, 0.5, (255,255,255), 1);
    cv2.putText(img, Texte4 ,(5,30), font, 0.5, (255,255,255), 1);
    cv2.putText(img, Texte ,(5,67), font, 0.5, (255,150,150), 1);
    #cv2.putText(img, Texte10 ,(5,440), font, 0.5, (255,255,255), 1);
    #cv2.putText(img, Texte19 ,(5,458), font, 0.5, (255,255,255), 1);
    #cv2.putText(img, Texte13 ,(5,470), font, 0.5, (255,255,255), 1);
    #cv2.putText(img, Texte12 ,(5,420), font, 0.5, (255,255,255), 1);
    ComptMin=str(Comptage_min)
    ComptMax=str(Comptage_max)
    ComptMoy=str(moyenne)
    texte5="min : "+ComptMin
    texte6="max : "+ComptMax
    texte7="moyenne : "+ComptMoy
    Texte14="x="+str(y2)
    Texte15="y="+str(x2)
    cv2.putText(img,texte5  ,(500,30), font, 0.5, (255,255,255), 1);
    cv2.putText(img, texte6 ,(500,60), font, 0.5, (255,255,255), 1);
    cv2.putText(img, texte7 ,(500,90), font, 0.5, (255,255,255), 1);
    cv2.putText(img, Texte14 ,(520,240), font, 0.5, (255,255,255), 1);
    cv2.putText(img, Texte15 ,(520,255), font, 0.5, (255,255,255), 1);
    ###############################
    #cv2.putText(img,"commandes:"  ,(5,94), font, 0.5, (255,255,255), 1);
    #cv2.putText(img,"___________" ,(5,97), font, 0.5, (255,255,255), 1);
    #cv2.putText(img, "q : Quitter" ,(5,120), font, 0.5, (255,255,255), 1);
    #cv2.putText(img, "m : isole / tapis" ,(5,140), font, 0.5, (255,255,255), 1);
    #cv2.putText(img, "c : couleur_milieu" ,(5,160), font, 0.5, (255,255,255), 1);
    #cv2.putText(img, "c : couleur_colonie" ,(5,180), font, 0.5, (255,255,255), 1);
    #cv2.putText(img, "e : eclaircir" ,(5,200), font, 0.5, (255,255,255), 1);
    #cv2.putText(img, "f : fonce" ,(5,220), font, 0.5, (255,255,255), 1);
    #cv2.putText(img, "a : augmente boite" ,(5,240), font, 0.5, (255,255,255), 1);
    #cv2.putText(img, "d : diminue boite" ,(5,260), font, 0.5, (255,255,255), 1);
    cv2.putText(img,"_______________________________________________" ,(0,40), font, 0.5, (255,150,150), 2);
    cv2.putText(img,"_______________________________________________" ,(0,8), font, 0.5, (255,150,150), 2);
    cv2.putText(img,"I" ,(422,38), font, 0.5, (255,150,150), 2);
    cv2.putText(img,"I" ,(422,28), font, 0.5, (255,150,150), 2);
    cv2.putText(img,"I" ,(422,18), font, 0.5, (255,150,150), 2);
    cv2.putText(img,"I" ,(245,38), font, 0.5, (255,150,150), 2);
    cv2.putText(img,"I" ,(245,28), font, 0.5, (255,150,150), 2);
    cv2.putText(img,"I" ,(245,18), font, 0.5, (255,150,150), 2);
    ####################
    if webi==1:
        zz=1
    else:
        cv2.putText(img,"___________________________________________" ,(250,475), font, 0.5, (155,250,150), 2);
        cv2.putText(img,"___________________________________________" ,(250,400), font, 0.5, (155,250,150), 2);
        cv2.putText(img,"I" ,(248,473), font, 0.5, (155,250,150), 2);
        cv2.putText(img,"I" ,(248,463), font, 0.5, (155,250,150), 2);
        cv2.putText(img,"I" ,(248,453), font, 0.5, (155,250,150), 2);
        cv2.putText(img,"I" ,(248,443), font, 0.5, (155,250,150), 2);
        cv2.putText(img,"I" ,(248,433), font, 0.5, (155,250,150), 2);
        cv2.putText(img,"I" ,(248,423), font, 0.5, (155,250,150), 2);
        cv2.putText(img,"I" ,(248,413), font, 0.5, (155,250,150), 2);
        texte40="Suggestions:"
        cv2.putText(img, texte40 ,(255,425), font, 0.5, (155,250,15), 1);
        cv2.putText(img, "___________" ,(255,435), font, 0.5, (255,150,150), 1);
        if tuto==0:
            texte41=". Calibration sur milieu"
            texte42=". Eclaircissement du milieu"
            cv2.putText(img, texte41 ,(380,425), font, 0.5, (155,250,15), 1);
            cv2.putText(img, texte42 ,(380,440), font, 0.5, (155,250,15), 1);
        if tuto==1:
            texte43=". Calibration sur colonie"
            texte44=". Cibler une bordure de colonie"
            texte45=". Ajustement des diametres"
            cv2.putText(img, texte43 ,(380,425), font, 0.5, (155,250,15), 1);
            cv2.putText(img, texte44 ,(380,440), font, 0.5, (155,250,15), 1);
            cv2.putText(img, texte45 ,(380,455), font, 0.5, (155,250,15), 1);
        if tuto==2:
            texte43=". Cibler une petite colonie"
            texte44=". Calibration non determinante"
            texte45=". Eclaircissement du milieu"
            cv2.putText(img, texte43 ,(380,425), font, 0.5, (155,250,15), 1);
            cv2.putText(img, texte44 ,(380,440), font, 0.5, (155,250,15), 1);
            cv2.putText(img, texte45 ,(380,455), font, 0.5, (155,250,15), 1);
        if tuto==3:
            texte43=". Calibration sur colonie"
            texte44=". Cibler une bordure de colonie"
            texte45="*Resolution des paquets :"
            #cv2.putText(img, "_____________________" ,(255,465), font, 0.5, (255,150,150), 1);
            texte46=" -->Modif. aire imposee  -->Mode paquet"
            cv2.putText(img, texte43 ,(380,425), font, 0.5, (155,250,15), 1);
            cv2.putText(img, texte44 ,(380,440), font, 0.5, (155,250,15), 1);
            cv2.putText(img, texte45 ,(255,455), font, 0.5, (155,250,15), 1);
            cv2.putText(img, texte46 ,(260,470), font, 0.5, (155,250,15), 1);
        if tuto==4:
            texte43=". Calibration sur colonie"
            texte44=". Cibler une bordure de colonie"
            texte45="*Resolution des paquets :"
            texte46=" -->Modif. aire imposee  -->Mode paquet"
            cv2.putText(img, texte43 ,(380,425), font, 0.5, (155,250,15), 1);
            cv2.putText(img, texte44 ,(380,440), font, 0.5, (155,250,15), 1);
            cv2.putText(img, texte45 ,(255,455), font, 0.5, (155,250,15), 1);
            cv2.putText(img, texte46 ,(260,470), font, 0.5, (155,250,15), 1);
        if tuto==5:
            texte43=". Calibration sur Milieu & xSmall"
            texte44=". Cibler une bordure de colonie"
            texte45=". Eclaircir a +7"
            cv2.putText(img, texte43 ,(380,425), font, 0.5, (155,250,15), 1);
            cv2.putText(img, texte44 ,(380,440), font, 0.5, (155,250,15), 1);
            cv2.putText(img, texte45 ,(380,455), font, 0.5, (155,250,15), 1);
        if tuto==6:
            texte43=". Calibration sur Milieu"
            texte44=". Cibler une bordure de colonie"
            texte45=". Eclaircir a +6"
            texte46=". Diametre sup a 5"
            cv2.putText(img, texte43 ,(380,425), font, 0.5, (155,250,15), 1);
            cv2.putText(img, texte44 ,(380,440), font, 0.5, (155,250,15), 1);
            cv2.putText(img, texte45 ,(380,455), font, 0.5, (155,250,15), 1);
            cv2.putText(img, texte46 ,(380,470), font, 0.5, (155,250,15), 1);
        if tuto==7:
            texte43=" A vous de jouer .."
            texte45="--> envoyer moi vos pickle !"
            texte46="       ( dfialaire@gmail.com )"
            cv2.putText(img, texte43 ,(380,425), font, 0.5, (155,250,15), 1);
            cv2.putText(img, texte45 ,(380,449), font, 0.5, (155,250,15), 1);
            cv2.putText(img, texte46 ,(380,470), font, 0.5, (215,120,0), 1);
        if tuto==8:
            texte43=" A vous de jouer .."
            texte45="--> envoyer moi vos 2 pickles (1/couleur)"
            texte46="       ( dfialaire@gmail.com )"
            cv2.putText(img, texte43 ,(380,425), font, 0.5, (155,250,15), 1);
            cv2.putText(img, texte45 ,(260,449), font, 0.5, (155,250,15), 1);
            cv2.putText(img, texte46 ,(380,470), font, 0.5, (215,120,0), 1);
        if tuto==9:
            texte43=" Ceci est votre capture de Webcam.."
            cv2.putText(img, texte43 ,(260,449), font, 0.5, (155,250,15), 1);
#######################""    

    cv2.imshow("En direct _ Video Webcam", img)
    cv2.moveWindow("En direct _ Video Webcam", 15,35)
    cv2.setMouseCallback("En direct _ Video Webcam",Souris)
    #####################################################       
    if(cv2.waitKey(1) ==ord("l")):
        print("ListeAirePaquet=",ListeAirePaquet)
        #print("Aire_col_imp=",Aire_col_imp)
        #print("diam_imp=",diam_imp)
        #print("SomeListeAirePaquet=sum(ListeAirePaquet)=",SomeListeAirePaquet)
        #print("SommColTot=SomeListeAirePaquet/Aire_col_imp=",SommColTot)
        
  
    #####################################################    
    if(cv2.waitKey(1) ==ord("p")):
        y3=input("Quelle valeur souhaitez vous pour la position x ? : ")
        y2=int(y3)
        x3=input("Quelle valeur souhaitez vous pour la position y ? : ")
        x2=int(x3)    
    #####################################################    
    if(cv2.waitKey(1) ==ord("6")):
        y2=y2+4
    #####################################################    
    if(cv2.waitKey(1) ==ord("4")):
        y2=y2-4
    #####################################################    
    if(cv2.waitKey(1) ==ord("8")):
        x2=x2-4
    #####################################################    
    if(cv2.waitKey(1) ==ord("2")):
        x2=x2+4
    #####################################################    
#####################################################     
## Nécessité de détruire la fenêtre Webcam !
#############################################
cam.release()
cv2.destroyAllWindows()
```
