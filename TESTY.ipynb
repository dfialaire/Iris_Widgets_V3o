```python
%%HTML
<button id="btn">
	begin gUM
</button><br>
<video id="video" autoplay controls muted playsinline></video>
<script>
const constraints = {
  video: {
    facingMode: {
      exact: "environment"
    }
  }
};
btn.onclick = e => {
	navigator.mediaDevices.getUserMedia(constraints)
		.then((stream) => {video.srcObject = stream})
		.catch( console.error );
};
</script>
```
